# Introduction to Computer Vision with Image Processing
In this section, we will explore the fundamental concepts and techniques in computer vision through image processing:

- Loading Images with OpenCV
- Estimating Image Dimensions
- Displaying and Viewing Images
- Grayscale Conversion
  - RGB Color Space
  - HSV Color Space
- Adjusting Brightness and Contrast


```python
# loading an image
import cv2
from google.colab.patches import cv_imshow
```

```python
# load image
img = cv2.imread("/content/dsn.png") # Change the file path to your path
```

```python
# image dimension
img.shape # (height, width, channel)
```

```python
# view and image with matplotlib
import matplotlib.pyplot as plt
plt.imshow(img) # View image with matplotlib
```

```python
# view and image

cv_imshow(img) # View image with cv_imshow
```

```python
# gray scaling
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv_imshow(gray_image)
```

```python
# rgb image
rgb_image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
cv_imshow(rgb_image)
```

```python
# hsv image
hsv_image = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
cv_imshow(hsv_image)
```

```python
# contrast / brightness control
# define the alpha and beta
alpha = 2.5 # Contrast control
beta = 20 # Brightness control

# call convertScaleAbs function
adjusted = cv2.convertScaleAbs(img, alpha=alpha, beta=beta) # Img, contrast and brightness
cv_imshow(adjusted)
```
